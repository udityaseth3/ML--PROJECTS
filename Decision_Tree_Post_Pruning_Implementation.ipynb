**Problem Statement:**

The problem statement here is to build a machine learning model that can accurately classify iris flowers into one of three species (setosa, versicolor, or virginica) based on their measurements (sepal length, sepal width, petal length, and petal width).
Identify the type of iris flower by looking at its physical characteristics. This is a supervised learning classification problem because we have labeled data (each flower has a known species), and we want to predict a categorical outcome (the species).

Description:

This project is a basic machine learning workflow using the Iris dataset. Here's a breakdown of the steps:

Load Data: The Iris dataset is loaded using scikit-learn and also using seaborn for visualization convenience.
Prepare Data: The dataset is split into features (X) and the target variable (y). The target variable represents the species of iris flower.
Split Data: The data is divided into training and testing sets to evaluate the model's performance on unseen data.
Train Model: A Decision Tree Classifier is trained on the training data. This model learns to classify iris flowers based on their features.
Visualize Tree: The trained decision tree is visualized to understand the decision rules it has learned.
Make Predictions: The trained model is used to predict the species of iris flowers in the test set.
Evaluate Model: The accuracy and classification report are calculated to assess how well the model performed on the test data.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
#independent feature and dependent features
X=df.iloc[:,:-1]
y=iris.target
```

```python
X,y
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## Postpruning
treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#prediction
y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
